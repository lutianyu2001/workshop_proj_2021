```python
year_match = all_event.groupby('year')['year'].count()
year_match
plt.figure(figsize=(10, 10))
plt.plot(year_match.index,year_match,label = 'number_of_marches')
plt.legend(loc='upper left',frameon=False)
plt.xlabel("year")
plt.ylabel("number_of_marches")
```

```python
friendly = all_event[all_event['tournament'] == 'Friendly']
year_match_fredly = friendly.groupby('year')['year'].count()
year_match_fredly
plt.figure(figsize=(10, 10))
plt.plot(year_match_fredly.index,year_match_fredly,label = 'number_of_marches')
plt.legend(loc='upper left',frameon=False)
plt.xlabel("year")
plt.ylabel("number_of_marches")
```

```python
plt.figure(figsize=(10, 10))
temp = std_fifa_ranking.sort_values(by="RNK")

plt.plot(temp['RNK'],temp['TOTAL_POINTS'],color = 'r',label = 'TOTAL_POINTS')
plt.plot(temp['RNK'],temp['PREVIOUS_POINTS'],color = 'blue',label = 'PREVIOUS_POINTS')
plt.legend(loc='upper right',frameon=False)
plt.xlabel("Rank")
plt.ylabel("Points")
```
